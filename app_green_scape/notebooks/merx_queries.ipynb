# **Configuración inicial**
---

```python
import mysql.connector
from sqlalchemy import create_engine
import pandas as pd

config = {
    'user': 'root',
    'password': 'root',
    'host': 'localhost', 
    'database': 'GreenScape',
    'port': 3306
}

def conectar_bd():
    try:
        conn = mysql.connector.connect(**config)
        return conn
    except mysql.connector.Error as err:
        print(f"Error de conexión: {err}")
        return None

print("Librerías cargadas.")
```

```python

conn = conectar_bd()
if conn:
    cursor = conn.cursor()
    cursor.execute("SHOW TABLES;")
    tablas = cursor.fetchall()
    
    print(f"Se encontraron {len(tablas)} tablas en el sistema:")
    print("-" * 30)
    for tabla in tablas:
        print(f"{tabla[0]}")
    
    cursor.close()
    conn.close()
```

```python
conn = conectar_bd()
engine = create_engine(f"mysql+mysqlconnector://{config['user']}:{config['password']}@{config['host']}:{config['port']}/{config['database']}")
```

```python
nombre_tabla="Clima"
query = f"DESCRIBE {nombre_tabla};"
df_usuario = pd.read_sql(query, con=engine)
print(f"Estructura de la tabla '{nombre_tabla}':")
display(df_usuario)
 
print("==="*50)
cursor = conn.cursor()
cursor.execute(f"SHOW CREATE TABLE {nombre_tabla};")
resultado = cursor.fetchone()
print(resultado[1]) 
```
